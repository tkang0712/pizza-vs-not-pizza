```python
from fastai.vision.all import *
```

```python
path = Path("/kaggle/input/pizza-not-pizza/pizza_not_pizza")
```

```python


dls = ImageDataLoaders.from_folder(path, valid_pct=0.2, item_tfms=Resize(224))

```

```python
dls.show_batch()
```

```python
learn = vision_learner(dls, resnet34, metrics=accuracy)
```

```python
learn.fine_tune(4)
```

```python
learn.show_results()
```

```python
learn.export('/kaggle/working/my_model.pkl')
```
